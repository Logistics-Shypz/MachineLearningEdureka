# Replacing NaN values with most frequent value in each column

```python
all_col = df.columns
```

```python
non_cat_col = df.describe().columns
```

```python
def Diff(li1,li2):
    return (list(set(li1)-set(li2)))
```

```python
cat_col_list = Diff(all_col,non_cat_col)
```

```python
cat_col_list
```

```python
def fillna(col):
    df.fillna(col.mode().iloc[0], inplace=True)
    return col
```

```python
df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
pd.isna(df)
```

# Fitting Decision tree Classifier and Calculating Accuracy

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
train,test = train_test_split(df,test_size=0.2,random_state=85)
```

```python
train.shape
```

```python
test.shape
```

```python
#len(cat_df.columns)
df.head(2)
```

```python
x_train = train.iloc[:,0:27]
x_test = test.iloc[:,0:27]
```

```python
y_train = train['outcome']
y_test = test['outcome']
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
#y_test
```

```python
print(metrics.accuracy_score(y_pred,y_test)*100,'%')
```

```python
y_pred
```

```python
import graphviz 
#dot_data = tree.export_graphviz(clf, out_file=None) 
dot_data = tree.export_graphviz(clf, out_file=None, 
                                    feature_names=x_test.columns,    
                                    filled=True, rounded=True,  
                                    special_characters=True)  
graph = graphviz.Source(dot_data) 
graph.render("horse") 
```
