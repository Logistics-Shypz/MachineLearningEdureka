```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Function to load a image and displaying it

```python
def load_image(img_name):
    im = Image.open(img_name)
    return im
#im = Image.open('dogs.jpeg')
```

```python
im = load_image('dogs.jpeg')
im
```

# Function to convert an Image to Numpy Array

```python
def img_to_array(im):
    data = np.asarray(im)
    return data
```

```python
dog_data = img_to_array(im)
```

```python
#dog_data
```

# Finding the dimensions of Image and converting to 2d

```python
width,height,channel = dog_data.shape
```

```python
#for i in range(0,len(dog_data)):
 #   print len(dog_data[i])
```

```python
dog_data_2d = dog_data.reshape(width*height,3)
```

```python
dog_data_2d.shape
```

```python
df = pd.DataFrame(dog_data_2d)
```

# Shuffling data frame 

```python
np.random.seed(100)
```

```python
from sklearn.utils import shuffle
shuffle_df = shuffle(df)
```

```python
shuffle_df.head()
```

```python
shuffle_df.describe()
```

```python
f1 = shuffle_df.iloc[:,0].values
f2 = shuffle_df.iloc[:,1].values
f3 = shuffle_df.iloc[:,2].values
```

```python
plt.scatter(f1,f2,f3,c='blue')
```

```python
shuffle_df.shape
```

# Applying K Means clustering algorithm

```python
from sklearn.cluster import KMeans
```

```python
#X = np.array(list(zip(f1,f2)))
```

```python
X = shuffle_df
```

```python
#X_train = X[0:52836]
#X_test = X[52836:]
X = np.array(list(zip(f1,f2,f3)))
X.shape
```

```python
kmeans_model = KMeans(n_clusters=7,random_state=83)
```

```python
kmeans_model.fit(X)
```

```python
kmeans_model.labels_
```

```python
kmeans_model.cluster_centers_
```

```python
cluster_centers = kmeans_model.cluster_centers_
```

```python
cluster_labels = kmeans_model.labels_
```

```python
plt.figure(figsize = (15,8))
#cluster_centers = cluster_centers.astype(np.uint8)
plt.imshow(cluster_centers[cluster_labels].reshape(width, height, 3))
```
